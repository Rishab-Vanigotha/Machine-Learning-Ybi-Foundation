```python
dict = {'PAYMENT': 0, 'TRANSFER' : 1, 'CASH_OUT':2, 'DEBIT':3, 'CASH_IN':4}
df['type']=df['type'].map(dict)
df.type[0:5]
```

```python
df['isFraud'] = df['isFraud'].map({0:'notFraud',1:'Fraud'})
df.head()
```

```python
x = data[['type','amount','oldbalanceOrg','newbalanceOrig']]
y= data['isFraud']
```

```python
from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest = train_test_split(x,y,test_size=0.3, random_state= 20)
```

```python
from sklearn.tree import DecisionTreeClassifier
cl = DecisionTreeClassifier()
model = cl.fit(xtrain,ytrain)
print(model.score(xtest,ytest))
```

```python
ypred = model.predict(xtest)
y_train_pred=model.predict(xtrain)
```

```python
from sklearn.metrics import accuracy_score,classification_report
print(accuracy_score(ytest,ypred))
print(accuracy_score(ytrain,y_train_pred))
```
